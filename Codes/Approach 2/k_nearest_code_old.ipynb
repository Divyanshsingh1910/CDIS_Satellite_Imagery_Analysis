```python
from geopy.distance import geodesic
```

```python
class k_nearest_class:

    def __init__(self, k , lat,long, dataset , POI):

        self.lat = lat
        self.long = long
        self. DOI = dataset
        self.k = k 
        self.Poi = POI
         

    def distance_points(self,lat1, long1):
        coords_1 = (self.lat, self.long)
        coords_2 = (lat1, long1)
        
        distance = geodesic(coords_1, coords_2).kilometers
        
        return distance
    
    def dataset_create(self):
        condition = (self.DOI['fclass'] == self.Poi)
        self.SDOI = self.DOI[condition]
        return self.SDOI

        
    def k_nearest(self):
        distances = []
      
        
        for index, row in self.SDOI.iterrows():
            dist = self.distance_points(row["Y-COORD"], row["X-COORD"] )
            distances.append((row['fclass'],row['name'], dist))
            
        nearest_points = sorted(distances, key=lambda x: x[2])[:self.k]
        return nearest_points




    
```

```python
given_lat = 25.15433548349917

given_long =  75.85818430365612

k = 5

dataset = df_POIS

POI = 'hotel'
```

```python
k_nearest_instance = k_nearest_class(k,given_lat, given_long, dataset, POI)

 
subset_dataset = k_nearest_instance.dataset_create()
 
nearest_points = k_nearest_instance.k_nearest()


for fclass,point, distance in nearest_points:
    print(f"Class: {fclass}",f"Point: {point}, Distance: {distance:.2f} kilometers")


```
