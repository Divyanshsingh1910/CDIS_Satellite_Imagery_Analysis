```python
column_names_set1 = set(df4.columns)
column_names_set2 = set(train_1.columns)
```

```python
# Calculate the number of common column names
common_columns = column_names_set1.intersection(column_names_set2)
num_common_columns = len(common_columns)

print(f"Number of common column names: {num_common_columns}")
print(f"Common column names: {common_columns}")
```

```python
df_train = train_1[common_columns]
```

```python
df_train
```

```python
column_heads = train_1.columns.tolist()
```

```python
column_heads
```

```python
df_5 = pd.DataFrame(0, columns=column_heads, index=range(98))
```

```python
df_5
```

```python
df_5.update(df4)
```

```python
concatenated_df = pd.concat([df_5, train_1], axis=0)
```

```python
concatenated_df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(concatenated_df)
concatenated_df_1 = pd.DataFrame(scaled_values, columns=concatenated_df.columns)
```

```python
X_test = concatenated_df_1.tail(98).copy()
```

```python
concatenated_df_1.drop(concatenated_df_1.tail(98).index, inplace=True)
```

```python
X_test
```

```python
concatenated_df_1
```

```python
X_train = concatenated_df_1
```

```python
X_train.fillna(0, inplace=True)
X_test.fillna(0, inplace=True)
y_train.fillna(0, inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfc = RandomForestRegressor()
rfc.fit(X_train, y_train)
y_rfc_predict = rfc.predict(X_test)
```

```python
y_rfc_predict
```

```python
numpy_array_y_test = y_test.values
```

```python
numpy_array_y_test
```

```python
import matplotlib.pyplot as plt
```

```python
correlation_coefficient = np.corrcoef(y_rfc_predict, numpy_array_y_test)[0, 1]
```

```python
plt.scatter(y_rfc_predict, numpy_array_y_test, label=f'Correlation: {correlation_coefficient:.2f}', color='b')
plt.xlabel('RWI_Prediction')
plt.ylabel('Income')
plt.legend()

plt.title('Scatter Plot of RWI_Prediction and Income')
plt.show()
```
