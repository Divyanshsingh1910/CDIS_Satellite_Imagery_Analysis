```python
from geopy.distance import geodesic
import geopandas as gpd
from shapely.geometry import Point
```

```python
df_NTL = pd.read_csv('/content/NTL data of haryana (1).csv')
df_llong = pd.read_csv('/content/lat-long_pred.csv')
```

```python
df_llong
```

```python
df_NTL.columns = ['Latitude', 'Longitude','NTL']
```

```python
df_NTL
```

```python
def calculate_distance(coord1, coord2):
    return geodesic(coord1, coord2).kilometers
```

```python
for i in range(98):
  given_latitude = df_llong.loc[i,'Latitude']
  given_longitude = df_llong.loc[i,'Longitude']

  given_coords = (given_latitude, given_longitude)
  df_NTL['Distance'] = df_NTL.apply(lambda row: calculate_distance(given_coords, (row['Latitude'], row['Longitude'])), axis=1)
  closest_index = df_NTL['Distance'].idxmin()

  closest_value = df_NTL.loc[closest_index, 'NTL']
  df_llong.loc[i,'NTL']=closest_value
  #print(f'The closest point has a value of {closest_value}')
  print(i)
  i=i+1
```
